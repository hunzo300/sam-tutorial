```python
import torch
import matplotlib.pyplot as plt
import os

def load_model_weights(path):
    return torch.load(path, map_location=torch.device('cpu'))

def plot_weight_distribution(model_weights, model_name):
    #   
    plt.figure(figsize=(10, 6))
    
    for name, weight in model_weights.items():
        if 'weight' in name:  #   (  )
            weight_flat = weight.flatten().cpu().numpy()  #  1D  
            plt.hist(weight_flat, bins=100, alpha=0.5, label=name)
    
    plt.title(f'Weight Distribution of {model_name}')
    plt.xlabel('Weight values')
    plt.ylabel('Frequency')
    plt.legend(loc='upper right')
    plt.show()

def compare_models_weights(path1, path2, model1_name, model2_name):
    #   
    model1_weights = load_model_weights(path1)
    model2_weights = load_model_weights(path2)
    
    #  1   
    plot_weight_distribution(model1_weights, model1_name)
    
    #  2   
    plot_weight_distribution(model2_weights, model2_name)

#    
sam_model_path = "/home/minkyukim/sam-tutorial/work_dir/SAM/sam_vit_b_01ec64.pth"
medsam_model_path = "/mnt/sda/minkyukim/pth/sam-tutorial_brats/medsam_model_best.pth"

#    
compare_models_weights(sam_model_path, medsam_model_path, "SAM-ViT-B", "MedSAM-ViT-B")
```
